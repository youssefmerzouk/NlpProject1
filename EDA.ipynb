```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
# Charger le dataset
file_path = 'sentences.csv'
data = pd.read_csv(file_path)
```

```python
data['darija_length'] = data['darija'].apply(len)
data['darija_ar_length'] = data['darija_ar'].apply(len)
data['eng_length'] = data['eng'].dropna().apply(len)
```

```python
#distribution des longueurs de texte
plt.figure(figsize=(10, 6))
plt.hist(data['darija_length'], bins=30, alpha=0.7, label='darija (Roman script)')
plt.hist(data['darija_ar_length'], bins=30, alpha=0.7, label='darija_ar (Arabic script)', color='orange')
plt.hist(data['eng_length'], bins=30, alpha=0.7, label='eng (English)', color='green')
plt.title('Distribution des longueurs des textes')
plt.xlabel('Longueur du texte')
plt.ylabel('Fréquence')
plt.legend()
plt.show()
```

```python
# Générer un WordCloud pour la colonne 'darija'
darija_text = ' '.join(data['darija'])
darija_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(darija_text)
```

```python
#afficher le WordCloud pour 'darija'
plt.figure(figsize=(10, 5))
plt.imshow(darija_wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WordCloud pour Darija (Roman script)')
plt.show()
```

```python
#générer un WordCloud pour la colonne 'darija_ar'
darija_ar_text = ' '.join(data['darija_ar'])
darija_ar_wordcloud = WordCloud(width=800, height=400, background_color='white', font_path=None).generate(darija_ar_text)
```
